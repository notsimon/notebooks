```python
%matplotlib inline

import numpy as np
from numpy import random as rd
from scipy import stats
import matplotlib as mpl
from matplotlib import pyplot as plt, rc
import seaborn as sns

sns.set_style("whitegrid")
```

This notebook shows how to sample from an arbitrary distribution.

## Inverse transform sampling

[Wikipedia article](http://en.wikipedia.org/wiki/Inverse_transform_sampling). In this case, we are sampling from a normal distribution `n`.

```python
u = stats.uniform(loc=0, scale=1)
n = stats.norm(loc=0, scale=1)

fig, ax = plt.subplots(1, 2, figsize=(14, 4), sharey=True)

ax[0].set_title('Uniform distribution')
ax[1].set_title('Normal distribution to sample from')

# distributions functions
for rv, p in zip([u, n], ax):
    x = np.linspace(rv.ppf(0.0001), rv.ppf(0.9999), 100)
    p.plot(x, rv.pdf(x), label='Probability density function')
    p.plot(x, rv.cdf(x), label='Cumulative distribution function')
    p.set_ylim(0, 1.4)

# random sample with its projection
rs = u.rvs() # input random value
x0, y0 = rs, u.cdf(rs)
x1, y1 = n.ppf(y0), y0 # output

ax[0].vlines(x0, 0, 1.4, linestyles='--', linewidth=1)
ax[1].vlines(x1, 0, 1.4, linestyles='--', linewidth=1)

tr = fig.transFigure.inverted()
p = np.array([tr.transform(ax[0].transData.transform((x0, y0))),
              tr.transform(ax[1].transData.transform((x1, y1)))])
l = mpl.lines.Line2D((p[:, 0]), (p[:, 1]), transform=fig.transFigure,
                     color='k', linewidth=1)
fig.lines.append(l)

ax[0].legend(loc='upper left');
```

Sampling from a discrete distribution.

```python
def cumulative(p):
    return np.array([sum(p[:i+1]) for i, __ in enumerate(p)])

def sample(pmf, size=1):
    cdf = cumulative(pmf)
    idx = np.zeros(size, dtype=int)
    
    for i in range(size):
        u = rd.random()
        n = np.nonzero(u < cdf)
        idx[i] = n[0][0]

    return idx
```

```python
x = np.linspace(0, 10, 10)
p_real = rd.random(size=10)
p_real /= sum(p_real)

s = sample(p_real, 4096)
p_samples, __ = np.histogram(s, bins=10)
p_samples = p_samples.astype(np.float) / sum(p_samples)

fig, ax = plt.subplots(1, 2, figsize=(14, 4))
ax[0].set_title('True distribution')
ax[1].set_title('Samples distribution')

col = sns.color_palette()

for i, p in enumerate([p_real, p_samples]):
    ax[i].bar(x, p, width=0.8, color=col[i], label='Probability mass function')
    ax[i].plot(x + 0.4, cumulative(p), color=col[i],
               label='Cumulative distribution function')
    ax[i].set_xlim(0, 12)
    ax[i].set_ylim(0, 1.2)
    ax[i].legend()
```
